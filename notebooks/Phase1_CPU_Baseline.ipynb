```python
# Run the CPU Training Executable on Colab CPU
import os
import pathlib

root_dir = pathlib.Path("..").resolve()
build_dir = root_dir / "build"

release_exe = build_dir / "cpu_train"
debug_exe = build_dir / "Debug" / "cpu_train.exe"
release_exe_win = build_dir / "Release" / "cpu_train.exe"

data_dir = root_dir / "data"

if release_exe.exists():
    # Linux-style executable (Colab typical case)
    !{release_exe} {data_dir}
elif release_exe_win.exists():
    # Windows-style Release exe (if present)
    !{release_exe_win} {data_dir}
elif debug_exe.exists():
    # Windows-style Debug exe (fallback)
    !{debug_exe} {data_dir}
else:
    print("cpu_train executable not found. Please check the build step above.")
```
